* Preliminary overview
This is a large dataset, with 1402844 rows and 24 columns. It contains 1 million + movies. I can see that the column headers already follow a snake_case naming convention, meaning I don't have to worry about implementing it.


Below are the column names, their data types and my additional notes on them:

* id (*dtype: integer*) - a unique identification key for each movie. **Keep this column**, as it could come in handy if I use another movie dataset.

* title (*dtype: object*) - 18 missing values - each movie's title. **Keep this column**.

* vote_average(*dtype: float*) - the average user rating, taken from the TMDB movie database. Could be interesting to do something with the general user popularity in my analysis. However, these particular average scores are from one site only and would not represent true general viewer ratings, in my opinion. If I could combine the data from this column with that of other datasets', which include user ratings from sites like IMDB and Rotten Tomatos, I could use it for my analysis. The issue is that right now, I'm exploring whether space missions affect the number of Sci-Fi movies that are made, **their popularity (in terms of people going to see the movie - box office metrics)**. Whether the film is any good or is enjoyed by the general audience is not a concern (**although audience reception could affect box office raitings, which would impact my analysis - I need the viewers' opinions during the first few weeks of release, as that would be the truest metric I could use**). **Delete this column and try to find more targeted formation**.


* vote_count (*dtype: integer*) - total count of viewer votes received. Informs the vote_average ratings. **Delete this column**.

* status (*dtype: object*) - the status of the movie (released, in production, etc.). **Investigate this and realease_date further**.

* release_date (*dtype: object*) - 306881 missing values - should be the release date of the movie. I say should, because **I've noticed a mismatch between the status and release_date columns**. Rumoured movies have release dates; some movies in post and pre - production don't have release dates. **Investigate further**

* revenue (*dtype: integer*) - revenue generated, **however it doesn't specify the currency and there are a lot of movies where the revenue is 0**. **Delete**; the overall revenue of a movie is not connected to my project.

* runtime (*dtype: integer*) - duration of the movie in minutes.

* adult (*dtype: bool*) - column has False if the movie is not suitable for children and True if it is. The logic here is a bit backwords in my opinion.

* backdrop_path (*dtype: object*) - 1052902 missing values - url for the movie backdrop photo. **Don't need it for my analysis - delete**.

* budget (*dtype: integer*) - shows the budget allocated for the movie. The column has no inforamtion however, about the currency these budgets are in.

* homepage (*dtype: object*) - 1258298 missing values - contains the URL for the movie's official website, for marketing purposes.

* imdb_id (*dtype: object*) - 737938 missing values - tells us what the film's ID is in IMDB's dataset.

* original_language (*dtype: object*) - the original language the film was produced in.

* original_title (*dtype: object*) - 18 missing values - the original title of the film (in it's native language).

* overview (*dtype: object*) - 320603 missing values - gives an overview of the movie's plot.

* popularity (*dtype: float*) - popularity score of the movie. There is no information howver on how that score is calculated; what is taken into account here or not; how it differs from the vote_average column.

* poster_path (*dtype: object*) - 495110 missing values - url for the movie poster photo. **Don't need it for my analysis - delete**.

* tagline (*dtype: object*) - 1207469 missing values - the movie's marketing tagline.

* genres (*dtype: object*) - 613360 missing values - the genre or genres of the movie.

* production_companies (*dtype: object*) - 803373 missing values - the production house/ houses that made the movie.

* production_countries (*dtype: object*) - 675619 missing values - countries that took part in the film's making (on-set or location shooting, pre and post production)

* spoken_languages (*dtype: object*) - 648632 missing values - what language's are spoken in the movie.

* keywords (*dtype: object*) -1054288 missing values -keywords, associated with the movie. No information of where they are taken from and the criteria of choosing them. Could be misleading.
