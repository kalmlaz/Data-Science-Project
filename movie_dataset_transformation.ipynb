Duplicate titles issue - there are 77641 duplicates in the title column, 2203 duplicates in the imdb column and 883 duplicates in the id field. I started cleaning the duplicates by subseting the three columns and removing rows where the data was identical. I then did the same, but I subsetted the three columns into groups of two, changing the combinations (title, id, imdb_id; then title and id; then id and imdb_id and so on). I managed to reduce the title column duplicates to 39852 and I was able to clean the id and imdb_id columns entirely. The id column was the first to be cleared of duplicates. The imdb_id column still had some instances of duplicate values however. I examined the rows where there were duplicates and found out that the duplicate imdb_id values were for different regional releases of a movie. So a movie released in the US (english), France (french), Spain (spanish) and so on would have a different id, but the same imbd_id. Another observation I made is that for some movies, a different release date in each country warrants a separate entry and id in the dataset (The imbd_id remained the same however). There were also some movies that were just duplicated due to a difference in spelling (be it by mistake or on purpose) or a differnce in the row information. I tried to consolidate the information, across the duplicate rows, as much as possible in Excel and then I removed the duplicates from the dataset. Only title duplicates remained with the majority being words or small sentences like 'Home', 'Mother', 'The Return', etc. After checking the titles I was able to find only one real duplicate which was 'Black Panther: Wakanda Forever' and I removed it.

For the status/ release_date columns - I first deleted all movie records before 1960, as they aren't a part of my analysis scope. I also removed rows where there's an NA value in the release_date and genres columns. I then checked the rows where the status is anything, except for 'Released. There were around 14 000 entries where the status was different from released and for almost all of them, there was a mismatch between their status and release_date (films are in production, rumored or in post-production, but have a release date < 2026). These mimatched entries start from row 358326, meaning that there are still thousands of correct film entries. A large part of these problematic films are foreign (outside of the US). I cannot check whether each of the planned or in production movies actually got released or not. 
I have deleted all of the 'Rumored' and 'Planned' entries, eliminating the biggest uncertainty factor from the dataset and allowing the data to be as certain as possible. This way I can more clearly test my hypothesis, without venturing into the realm of making predictions or relying on speculated trends. There were also some films with a "Canceled" status. which I also deleted. In the end I also decided to delete the 'In production' entries and the 'Post Production' ones. My reasoning for this is - a film can begin production, but end up in production hell where it is never finished - either due to money running out, not enough man power or other countless problems arising. The same can happen in post production. Seeing as I don't want to speculate or make any future predictions, I stuck only with the released movies.

For the production_countries column - changed United States of America to US and United Kingdom to UK, so it's the same as the space missions dataset.

```python
#Cleaning the dataset. 
#Removing the uneeded columns.
film_dataset_adjusted = film_dataset.drop(columns = ["vote_average", "vote_count", "revenue", "runtime", "adult", "backdrop_path", "budget", "homepage", "popularity", "poster_path", "spoken_languages"])

#Change the values in the release_date column from objects to real dates.
film_dataset_adjusted['release_date'] = pd.to_datetime(film_dataset_adjusted['release_date'], errors='coerce')

#Delete rows with release_date before 1st January 1960.
film_dataset_adjusted = film_dataset_adjusted[film_dataset_adjusted['release_date'] >= '1960-01-01']

#Remove the rows which have an NA value in the release_date column.
film_dataset_adjusted = film_dataset_adjusted.dropna(subset=['release_date'])

film_dataset_adjusted = film_dataset_adjusted.dropna(subset=['genres'])

#Process of cleaning duplicate movies.
#Remove rows where there are duplicate title, imdb and id values (only combinations of all three), while keeping the firs instance of that combination. 
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'imdb_id', 'id'],
    keep='first',
    inplace=True)

#Remove rows where there are duplicate title and imdb values (only combinations of those two), while keeping the firs instance of that combination.
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'imdb_id'],
    keep='first',
    inplace=True)

#Remove rows where there are duplicate title and id values (only combinations of those two), while keeping the firs instance of that combination.
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'id'],
    keep='first',
    inplace=True)

#Remove the remaining duplicate movie, by referencing its row
film_dataset_adjusted = film_dataset_adjusted.drop(799201)

#Remove rows where there are duplicate imdb_id values, while keeping the firs instance they appear in.
film_dataset_adjusted.drop_duplicates(
    subset=['imdb_id'],
    keep='first',
    inplace=True)

#Remove the rows where the status is either "Rumored" or 'Planned" in order to manage uncertainty in the dataset
film_dataset_adjusted = film_dataset_adjusted[~film_dataset_adjusted["status"].isin(["Rumored", "Planned", "Canceled", "In Production", "Post Production"])]

#Changes the country names for the US and the UK
film_dataset_adjusted["production_countries"] = (film_dataset_adjusted["production_countries"].str.replace("United Kingdom", "UK", regex=False).str.replace("United States of America", "USA", regex=False))
```
