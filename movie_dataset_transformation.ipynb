```python
#First round of cleaning for the dataset. 
#Removing the uneeded columns.
film_dataset_adjusted = film_dataset.drop(columns = ["vote_average", "vote_count", "revenue", "runtime", "adult", "backdrop_path", "budget", "homepage", "popularity", "poster_path", "spoken_languages"])

#Change the values in the release_date column from objects to real dates.
film_dataset_adjusted['release_date'] = pd.to_datetime(film_dataset_adjusted['release_date'], errors='coerce')

#Delete rows with release_date before 1st Janusary 1960.
film_dataset_adjusted = film_dataset_adjusted[film_dataset_adjusted['release_date'] >= '1960-01-01']

#Remove the rows which have an NA value in the release_date column.
film_dataset_adjusted = film_dataset_adjusted.dropna(subset=['release_date'])

film_dataset_adjusted = film_dataset_adjusted.dropna(subset=['genres'])

#Process of cleaning duplicate movies.
#Remove rows where there are duplicate title, imdb and id values (only combinations of all three), while keeping the firs instance of that combination. 
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'imdb_id', 'id'],
    keep='first',
    inplace=True)

#Remove rows where there are duplicate title and imdb values (only combinations of those two), while keeping the firs instance of that combination.
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'imdb_id'],
    keep='first',
    inplace=True)

#Remove rows where there are duplicate title and id values (only combinations of those two), while keeping the firs instance of that combination.
film_dataset_adjusted.drop_duplicates(
    subset=['title', 'id'],
    keep='first',
    inplace=True)
```
