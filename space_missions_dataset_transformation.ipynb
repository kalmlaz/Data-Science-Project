```python
# Logic to check the column's data types
space_missions_numeric = space_missions.select_dtypes(include=["number"]).columns.tolist()
space_missions_obj_cat = space_missions.select_dtypes(include=["object", "category"]).columns.tolist()
space_missions_bool = space_missions.select_dtypes(include=["bool"]).columns.tolist()

# Print results
print(f"Columns with numeric data are: {space_missions_numeric}")
print(f"Columns with categorical data are: {space_missions_obj_cat}")
print(f"Columns with boolean data are: {space_missions_bool}")
```

While the majority of the Datetime column's values are like - Wed Feb 05, 1958 07:33 UTC - there are around 15+ cells which provide less information - Sun Jun 26, 2022 (they are missing the time and UTC bit). This lack of a uniform pattern, in the text, along with a later realization that there are cells whith additional spaces, presented a challenge. Ultimately, a new column was created, that had the required date formating.
For the locations, each row had different text length, some had specific regions mentioned or states. I tried to turnicate the text, but the output wasn't what I needed. I then checked the turnicated values for duplicates, created a list with the different countries and used it to achive a column that contained only the country names.

```python
#Altering the original dataset under a new variable, so that I don't lose the first one.
#Removed the unnecessary columns
space_missions_modified = space_missions.drop(columns = ["Unnamed: 0", "Price", "Rocket_Status"])

#Transforming the dates data
# This creates a new column and copies the day and month abbreviations, date, comma and year. It searches for spaces, digit:digit, UTC, spaces 
#and deletes them.
space_missions_modified["date"] = space_missions_modified["Datetime"].str.replace(
    r"\s+\d{2}:\d{2}\s+UTC$", "", regex=True)

#The new date column in turned into actual date-time object (they were strings before). The errors bit is used so that if a cell can't be converted, it 
#becomes NaT (not a time) -> to check if there are any NaT's - space_missions_modified["date"].isna().any(). 
space_missions_modified["date"] = pd.to_datetime(space_missions_modified["date"], errors="coerce")

#Changes how the dates are displayed.
space_missions_modified["date"] = space_missions_modified["date"].dt.strftime("%d/%m/%Y")


#Removes the original Datetime column.
space_missions_modified.drop(columns = ["Datetime"], inplace = True)

#Changes the column titles
space_missions_modified.rename(columns={"Organisation": "organisation", "Location": "country", "Details": "mission_details", "Mission_Status": "mission_status"}, inplace = True)


space_missions_modified
```
